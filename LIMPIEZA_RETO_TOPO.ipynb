```python
id_map = df_all[df_all['Id Conductor'].notnull()].drop_duplicates(subset='Conductor') \
         .set_index('Conductor')['Id Conductor'].to_dict()
```

```python
df_all['Id Conductor'] = df_all.apply(
    lambda row: id_map[row['Conductor']] if pd.isnull(row['Id Conductor']) and row['Conductor'] in id_map else row['Id Conductor'],
    axis=1
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Variables a graficar
variables = ['KG C02', 'Rendimiento', 'Recorrido', 'Cantidad Mercancía']

# Crear figura
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(12, 12))
fig.suptitle('Boxplot y Distribución por Variable (Muestra de 1000)', fontsize=16)

for i, var in enumerate(variables):
    # Filtrar nulos y muestrear hasta 1000 datos
    data = df_all[var].dropna().sample(n=min(1000, df_all[var].dropna().shape[0]), random_state=42)

    # Boxplot (columna izquierda)
    sns.boxplot(x=data, ax=axes[i, 0], color='skyblue')
    axes[i, 0].set_title(f'Boxplot de {var}')
    axes[i, 0].set_xlabel('')
    axes[i, 0].set_yticks([])

    # Histograma (columna derecha)
    sns.histplot(data, bins=30, kde=False, ax=axes[i, 1], color='salmon', element="step")
    axes[i, 1].set_title(f'Distribución de {var}')
    axes[i, 1].set_xlabel('')

# Ajustes finales
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns
numeric_df = df_all.select_dtypes(include='number')

# Compute correlation matrix
corr = numeric_df.corr().round(2)

# Set up the matplotlib figure
plt.figure(figsize=(14, 10))
plt.title("Matriz de Correlación", fontsize=16, pad=20)

# Draw the heatmap
sns.heatmap(
    corr,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    linewidths=0.5,
    linecolor='white',
    square=True,
    cbar_kws={"shrink": 0.75}
)

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Variables a graficar
categorical_vars = ['Id Región', 'Cliente', 'Month', 'BL', 'Division']

# Configura figura
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16, 10))
fig.suptitle('Gráficos de Pastel por Variable Categórica (>3%)', fontsize=18, y=1.02)

# Aplanamos los ejes para fácil iteración
axes = axes.flatten()

# Función para mostrar porcentaje y etiqueta solo si >3%
def autopct_format(pct, allvals):
    absolute = int(round(pct/100.*sum(allvals)))
    return f'{pct:.1f}%\n' if pct > 3 else ''

# Graficar cada pie chart
for i, var in enumerate(categorical_vars):
    value_counts = df_all[var].value_counts()
    values = value_counts.values
    labels = [label if (v / sum(values)) > 0.03 else '' for label, v in zip(value_counts.index, values)]

    wedges, texts, autotexts = axes[i].pie(
        values,
        labels=labels,
        autopct=lambda pct: autopct_format(pct, values),
        startangle=90
    )
    axes[i].set_title(var, fontsize=12)

# Ocultar ejes vacíos si hay menos de 6 variables
for j in range(len(categorical_vars), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
